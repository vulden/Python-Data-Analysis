```python
data = pd.DataFrame(columns = ['method','nan_rate','precision_0', 'recall_0','support_0','precision_1', 'recall_1','support_1'])
```

```python
X = df.iloc[:, :-1].values  
y = df.iloc[:, 23].values  
X_imp = X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import sklearn.impute
from sklearn.decomposition import PCA
```

```python
# без удалений, запустим модели
model = RandomForestClassifier(random_state=42,
                               # число деревьев в лесу
                               n_estimators=80,
                               # функция для impurity ('gini' или 'entropy')
                               criterion='gini',
                               # то ли число слоев, то ли число расщеплений  
                               #  ?????????????????  но не число узлов
                               max_depth=5,
                               # сколько переменных использовать при построении дерева?
                               # sqrt(k)
                               max_features='auto', 
                               # максимальное число конечных узлов. Сейчас - без ограничений
                               max_leaf_nodes=None,
                               # минимальное уменьшение загрязнения 
                               min_impurity_decrease=0.0, 
                               # Устарело
                               min_impurity_split=None,
                               # Дерево обучается по части наблюдений (По какой???)
                               bootstrap=True, 
                               # минимальное число наблюдений в узле потомке
                               min_samples_leaf=1, 
                               # минимальное число наблюдений в узле родителе
                               min_samples_split=2,
                               # параметр используется редко, самостоятельно
                               min_weight_fraction_leaf=0.0, 
                               # объем промежуточных результатов в ходе обучения
                               verbose=1, 
                               # число используемых процессоров
                               # n_jobs=None,
                               # Вычислять out-of-bag ошибку
                               oob_score=True,
                               # использовать результаты предыдущего вызова и нарастить предыдущий лес 
                               warm_start=False,
                               # веса классов для балансировки обучения
                               class_weight=None)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)  
model.fit(X_train, y_train)
y_pred_train3 = model.predict(X_train)
y_pred_test3 = model.predict(X_test)
res = metrics.classification_report(y_pred_test3, y_test,output_dict=True)
data.loc[0] = ['full',0,res.get('0.0').get('precision'),res.get('0.0').get('recall'),res.get('0.0').get('f1-score'),res.get('1.0').get('precision'),res.get('1.0').get('recall'),res.get('1.0').get('f1-score')]
```

```python
nanrate = [0.1,1/4,1/3]
```

```python
random.seed(42)
for j in range(3):
    X_imp = X.copy()
    for i in range(int(nanrate[j]*X.size)):
        t2=random.randint(0, X.shape[1]-1)
        t1=random.randint(0, X.shape[0]-1)
        X_imp[t1,t2]=np.nan
    imp_mean = sklearn.impute.SimpleImputer(missing_values=np.nan,strategy="mean")
    imp_mean.fit(X_imp)
    S = imp_mean.transform(X_imp)
    X_train, X_test, y_train, y_test = train_test_split(S, y, test_size=0.2, random_state=1234)
    model.fit(X_train, y_train)
    y_pred_test3 = model.predict(X_test)
    res = metrics.classification_report(y_pred_test3, y_test,output_dict=True)
    data.loc[j+1] = ['mean',nanrate[j],res.get('0.0').get('precision'),res.get('0.0').get('recall'),res.get('0.0').get('f1-score'),res.get('1.0').get('precision'),res.get('1.0').get('recall'),res.get('1.0').get('f1-score')]
```

```python
X_scaled = preprocessing.StandardScaler().fit_transform(X)
```

```python
random.seed(42)
for j in range(3):
    X_imp = X_scaled.copy()
    for i in range(int(nanrate[j]*X.size)):
        t2=random.randint(0, X.shape[1]-1)
        t1=random.randint(0, X.shape[0]-1)
        X_imp[t1,t2]=np.nan
    L =[X_imp.copy(),X_imp.copy(),X_imp.copy()]
    vals = [3,5,7]
    S = []
    test_acc = []
    for t in range(3):
        imp_mean = sklearn.impute.KNNImputer(missing_values=np.nan,n_neighbors=vals[t])
        imp_mean.fit(L[t])
        S.append(imp_mean.transform(L[t]))
        X_train, X_test, y_train, y_test = train_test_split(S[t], y, test_size=0.2, random_state=1234)
        model.fit(X_train, y_train)
        test_acc.append(model.score(X_test,y_test))
    best = np.argmax(test_acc)
    X_train, X_test, y_train, y_test = train_test_split(S[best], y, test_size=0.2, random_state=1234)
    model.fit(X_train, y_train)
    y_pred_test3 = model.predict(X_test)
    res = metrics.classification_report(y_pred_test3, y_test,output_dict=True)
    data.loc[j+4] = ['KNN',nanrate[j],res.get('0.0').get('precision'),res.get('0.0').get('recall'),res.get('0.0').get('f1-score'),res.get('1.0').get('precision'),res.get('1.0').get('recall'),res.get('1.0').get('f1-score')]
```
